```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autoviml.Auto_NLP import Auto_NLP
```

```python
data = pd.read_csv("train.csv", encoding='ISO-8859-1', engine='python')
```

```python
data.shape
```

```python
data.head(2)
```

```python
train_data, test_data = train_test_split(data, test_size=0.2)
```

```python
# We have to tell AutoNLP that we are having a classifying problem
# if top_num_features not set then the default value more than 300 will take which is much slower
train_x, test_x, final, predicted= Auto_NLP(nlp_column='SentimentText', 
                                            train=train_data, 
                                            test=test_data,
                                            target='Sentiment',
                                            score_type="balanced_accuracy",
                                            top_num_features=100,
                                            modeltype="Classification", 
                                            verbose=2, 
                                            build_model=True)
```
